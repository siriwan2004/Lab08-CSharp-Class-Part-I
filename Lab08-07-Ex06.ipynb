## คำถาม

1. Method with parameter คืออะไร


*คำตอบ*

        เมธอด (หรือฟังก์ชัน) ที่รับค่าตัวแปร (parameters) เข้ามาเพื่อใช้ในการประมวลผลภายในเมธอดนั้น ๆ

2. ทำไมเราต้องใช้ parameter ใน method


*คำตอบ*

        การใช้พารามิเตอร์ในเมธอดช่วยให้เมธอดนั้นมีความยืดหยุ่นและสามารถทำงานกับข้อมูลที่แตกต่างกันได้ โดยไม่จำเป็นต้องเขียนเมธอดใหม่สำหรับข้อมูลแต่ละชุด

3. parameter และ argument แตกต่างกันอย่างไร


*คำตอบ*

Parameter: ตัวแปรที่ถูกประกาศในส่วนหัวของเมธอด ใช้เป็นตัวแทนของค่าที่จะถูกส่งเข้ามาเมื่อเรียกใช้เมธอด

Argument: ค่าจริงที่ถูกส่งเข้าไปยังเมธอดเมื่อเรียกใช้เมธอด

4. ส่วนประกอบของ method ที่มี parameter มีอะไรบ้าง


*คำตอบ*

Access Modifier: กำหนดระดับการเข้าถึงของเมธอด เช่น public, private

Return Type: ชนิดข้อมูลที่เมธอดจะส่งกลับ เช่น void, int, string

Method Name: ชื่อของเมธอด

Parameters: ตัวแปรที่รับค่าจากผู้เรียกใช้เมธอด

Method Body: โค้ดที่ทำงานภายในเมธอด

5. เมื่อไหร่ควรใช้ method ที่มี parameter

*คำตอบ*

        ควรใช้เมธอดที่มีพารามิเตอร์เมื่อคุณต้องการให้เมธอดนั้นทำงานกับข้อมูลที่แตกต่างกัน โดยไม่ต้องเขียนเมธอดใหม่สำหรับแต่ละชุดข้อมูล ซึ่งช่วยลดการทำงานซ้ำซ้อนและทำให้โค้ดของคุณมีความยืดหยุ่นมากขึ้น

8. จงเขียน method ที่ตรวจสอบว่าตัวเลขที่รับเข้ามาเป็นจำนวนคู่หรือจำนวนคี่ และแสดงผลลัพธ์ทางหน้าจอ

````
// พื้นที่เขียน code และทดสอบ
var mc = new MyClass();
mc.DisplayParam(99, 28); // ทดสอบฟังก์ชัน

class MyClass
{
    public void DisplayParam(int a, int b)
    {
        System.Console.WriteLine($"a = {a}, {CheckEvenOdd(a)}");
        System.Console.WriteLine($"b = {b}, {CheckEvenOdd(b)}");
    } 

    private string CheckEvenOdd(int num)
    {
        return (num % 2 == 0) ? "Even" : "Odd";
    }
}

````
