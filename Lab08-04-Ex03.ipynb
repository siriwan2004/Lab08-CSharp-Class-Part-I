## คำถาม

1. Access modifier ในภาษา C# คืออะไร


*คำตอบ*

        คือคีย์เวิร์ดที่ใช้กำหนดระดับการเข้าถึงของคลาส เมธอด ฟิลด์ และสมาชิกอื่น ๆ เพื่อควบคุมการมองเห็นและการเข้าถึงจากโค้ดภายนอก การใช้ Access Modifiers ช่วยในการซ่อนรายละเอียดการทำงานภายใน (encapsulation) และป้องกันการเข้าถึงหรือแก้ไขข้อมูลโดยไม่ได้ตั้งใจ

2. จงอธิบายขอบเขตและการใช้งานของ Access modifier ต่อไปนี้ 


2.1  public


*คำตอบ*

        สมาชิกหรือคลาสที่ถูกกำหนดเป็น public สามารถเข้าถึงได้จากทุกที่ ไม่ว่าจะเป็นภายในหรือภายนอกแอสเซมบลี

2.2 private


*คำตอบ*

        สมาชิกที่ถูกกำหนดเป็น private สามารถเข้าถึงได้เฉพาะภายในคลาสที่ประกาศเท่านั้น

2.3 protected


*คำตอบ*

        สมาชิกที่ถูกกำหนดเป็น protected สามารถเข้าถึงได้ภายในคลาสที่ประกาศและคลาสที่สืบทอด (derived classes)

2.4 internal


*คำตอบ*

        สมาชิกหรือคลาสที่ถูกกำหนดเป็น internal สามารถเข้าถึงได้เฉพาะภายในแอสเซมบลีเดียวกันเท่านั้น

2.5 protected internal

*คำตอบ*

        สมาชิกที่ถูกกำหนดเป็น protected internal สามารถเข้าถึงได้ภายในแอสเซมบลีเดียวกัน หรือจากคลาสที่สืบทอดในแอสเซมบลีอื่น
