## คำถาม

1. static class คืออะไร และแตกต่างจาก class ธรรมดาอย่างไร

*คำตอบ*

static class คือคลาสที่ไม่สามารถสร้างอ็อบเจกต์ (instance) ของมันได้ และสามารถมีเฉพาะสมาชิกที่เป็น static เท่านั้น (เช่น ฟังก์ชันหรือฟิลด์ที่เป็น static) แตกต่างจาก class ธรรมดา:
class ธรรมดา สามารถสร้างอ็อบเจกต์จากคลาสนั้นๆ ได้ และสามารถมีสมาชิกที่เป็นทั้ง static และ non-static
static class จะมีสมาชิกทั้งหมดเป็น static และไม่สามารถสร้างอ็อบเจกต์จากคลาสนี้ได้

2. เมื่อไหร่ควรใช้ static class และเมื่อไหร่ควรใช้ class ธรรมดา

*คำตอบ*

ใช้ static class เมื่อ:
คุณต้องการคลาสที่มีเฉพาะเมธอดหรือฟังก์ชันที่ไม่ขึ้นกับสถานะ (state) ของอ็อบเจกต์ เช่น เครื่องมือที่ใช้สำหรับคำนวณ, ฟังก์ชันที่ไม่เกี่ยวข้องกับข้อมูลเฉพาะ.
ตัวอย่าง: คลาสที่มีฟังก์ชันการคำนวณทั่วไปหรือฟังก์ชันทางคณิตศาสตร์ เช่น Math ใน C#

ใช้ class ธรรมดาเมื่อ:
คุณต้องการคลาสที่สามารถมีสถานะ (state) และต้องการสร้างหลายอ็อบเจกต์จากคลาสนี้.
ตัวอย่าง: คลาสที่ใช้ในการสร้างอ็อบเจกต์ที่มีข้อมูลที่ไม่เหมือนกัน เช่น Car, Person, หรือ Customer

3. สมาชิกทั้งหมดของ static class ต้องเป็น static หรือไม่ เพราะอะไร

*คำตอบ*

สมาชิกทั้งหมดใน static class ต้องเป็น static
เนื่องจาก static class ไม่สามารถสร้างอ็อบเจกต์จากคลาสนั้นๆ ได้ และสมาชิกที่ไม่เป็น static จะต้องเกี่ยวข้องกับอ็อบเจกต์ของคลาส ดังนั้น static class ต้องมีสมาชิกที่ไม่ขึ้นกับอ็อบเจกต์หรือสถานะ

4. เราสามารถสร้าง instance ของ static class ได้หรือไม่  เพราะอะไร

*คำตอบ*

ไม่สามารถสร้าง instance ของ static class ได้
เนื่องจาก static class ถูกออกแบบมาเพื่อใช้งานโดยไม่ต้องสร้างอ็อบเจกต์ ดังนั้นมันไม่สามารถมีตัวตน (instance) ที่ถูกสร้างขึ้นมาได้ ทุกสมาชิกใน static class ต้องสามารถเข้าถึงได้โดยตรงจากคลาสโดยไม่ต้องสร้างอ็อบเจกต์
