## คำถาม



1. Class ในภาษา C# คืออะไร


*คำตอบ*

        Class ในภาษา C# คือ โครงสร้างหรือแม่แบบที่ใช้ในการสร้างอ็อบเจ็กต์ (Object) โดยกำหนดคุณสมบัติ (Properties) และพฤติกรรม (Methods) ของอ็อบเจ็กต์นั้น ๆ คำว่า "Class" เป็นการกำหนดลักษณะของอ็อบเจ็กต์ ซึ่งจะทำให้เราสามารถสร้างอ็อบเจ็กต์จาก Class นี้ได้มากมายตามที่ต้องการ

2. Class มีหน้าที่อะไร


*คำตอบ*

        หน้าที่หลักของ Class คือการกำหนดรูปแบบของอ็อบเจ็กต์ และเป็นที่เก็บข้อมูลและฟังก์ชันต่าง ๆ ที่อ็อบเจ็กต์นั้นจะใช้ โดยการใช้ Class ช่วยให้การจัดการข้อมูลและฟังก์ชันเป็นระเบียบและสามารถนำกลับมาใช้ซ้ำได้ (Reusability)

3. ทำไมเราต้องใช้ Class


*คำตอบ*

        การใช้ Class ช่วยให้การพัฒนาโปรแกรมมีความเป็นระเบียบและยืดหยุ่นมากขึ้น ทำให้สามารถสร้างอ็อบเจ็กต์ที่มีคุณสมบัติและพฤติกรรมที่เหมือนกันได้หลายตัว Class ยังช่วยให้ง่ายต่อการบำรุงรักษาและขยายโปรแกรมในอนาคต เช่น การจัดกลุ่มข้อมูลและฟังก์ชันที่เกี่ยวข้องกันในที่เดียว

4. ส่วนประกอบของ Class มีอะไรบ้าง


*คำตอบ*

ส่วนประกอบของ Class ประกอบไปด้วย:

Fields: ตัวแปรที่ใช้ในการเก็บข้อมูลภายใน Class

Properties: การเข้าถึงข้อมูลภายใน Class โดยให้สามารถอ่านและเขียนค่าได้

Methods: ฟังก์ชันที่ใช้ดำเนินการหรือเปลี่ยนแปลงข้อมูลใน Class

Constructors: ฟังก์ชันที่ใช้ในการสร้างอ็อบเจ็กต์จาก Class

Destructors: ฟังก์ชันที่ใช้ในการทำความสะอาดทรัพยากรเมื่อต้องการทำลายอ็อบเจ็กต์

Events: การระบุเหตุการณ์ที่สามารถส่งสัญญาณให้กับอ็อบเจ็กต์อื่น ๆ ได้

5. Class และ Object แตกต่างกันอย่างไร


*คำตอบ*

Class คือ แม่แบบหรือโครงสร้างที่ใช้ในการกำหนดรูปแบบของข้อมูลและพฤติกรรมที่อ็อบเจ็กต์ (Object) จะมี ตัว Class จะเป็นตัวกำหนดคุณสมบัติ (Properties) และฟังก์ชัน (Methods) ของอ็อบเจ็กต์ และมันไม่ได้เก็บข้อมูลใด ๆ จริง ๆ แต่เป็นเพียงแค่การออกแบบหรือ Blueprint สำหรับการสร้างอ็อบเจ็กต์

Object คือ ตัวอย่างที่สร้างขึ้นจาก Class ซึ่งมีข้อมูลจริง (ค่า) และสามารถดำเนินการได้ตามฟังก์ชันที่ Class กำหนดไว้ เมื่อเราใช้คำสั่ง new เพื่อสร้าง Object จาก Class จะได้ตัวอ็อบเจ็กต์ที่มีข้อมูลจริงที่สามารถใช้งานได้

 Class เป็นการกำหนดรูปแบบ ส่วน Object คือสิ่งที่ถูกสร้างขึ้นจาก Class ซึ่งมีข้อมูลและสามารถใช้งานได้จริง

6. การตั้งชื่อ Class มีหลักเกณฑ์อย่างไร

*คำตอบ*

การตั้งชื่อ Class ในภาษา C# มีหลักเกณฑ์ดังนี้:

1.ใช้ตัวอักษรตัวใหญ่เริ่มต้น: ชื่อต้องเริ่มต้นด้วยตัวอักษรตัวใหญ่ (PascalCase) เช่น Person, CarModel

2.ตั้งชื่อให้สื่อความหมาย: ชื่อ Class ควรสะท้อนถึงสิ่งที่มันแทนหรือทำงาน เช่น Student, Employee, OrderManager

3.หลีกเลี่ยงการใช้ตัวเลข: ไม่ควรตั้งชื่อที่มีตัวเลขนำ เช่น 2DShape ควรใช้ชื่อที่เป็นคำแทน เช่น Shape2D

4.ไม่ใช้คำสำรอง (Reserved Keywords): หลีกเลี่ยงการใช้คำที่เป็นคีย์เวิร์ดในภาษา C# เช่น class, void, int

5.ใช้ชื่อที่ไม่ยาวเกินไป: ชื่อ Class ควรกระชับและตรงประเด็น โดยไม่ควรยาวเกินไป เช่น ใช้ Car แทน VehicleForTransportation

6.ไม่ใช้ตัวอักษรพิเศษหรือเว้นวรรค: ชื่อ Class ต้องประกอบด้วยตัวอักษร, ตัวเลข (ไม่ใช้ตัวเลขที่ขึ้นต้น) และเครื่องหมายขีดล่าง (_) หากจำเป็น

7. คำสั่ง p.GetType() คืออะไร

*คำตอบ*

        ใช้เพื่อเรียกดูชนิด (Type) ของอ็อบเจ็กต์ที่ถูกเก็บในตัวแปร p ซึ่งจะคืนค่าของชนิดของอ็อบเจ็กต์ในรูปแบบของ Type object ที่ใช้ในการจัดการข้อมูลเกี่ยวกับประเภทของอ็อบเจ็กต์นั้น ๆ

8. ผลที่ได้จากคำสั่ง  p.GetType() คืออะไร มันให้ข้อมูลอะไรแก่เราบ้าง

*คำตอบ*

คำสั่ง p.GetType() จะคืนค่าของชนิด (Type) ของอ็อบเจ็กต์ที่อยู่ในตัวแปร p ซึ่งผลลัพธ์ที่ได้จะเป็นอ็อบเจ็กต์ชนิด Type ที่แสดงถึงข้อมูลเกี่ยวกับประเภท (Type) ของอ็อบเจ็กต์นั้น ๆ

ข้อมูลที่สามารถได้จาก p.GetType() ได้แก่:

        ชื่อของชนิด (Type Name): จะบอกชื่อคลาสหรือชนิดของอ็อบเจ็กต์ที่ p แทน เช่น ถ้า p เป็นอ็อบเจ็กต์ของคลาส Car ผลลัพธ์จาก p.GetType() จะคืนชื่อชนิดเป็น Car

        ข้อมูลเกี่ยวกับชนิด (Type Information): ผ่านอ็อบเจ็กต์ Type ที่ได้จาก GetType() เราสามารถเข้าถึงข้อมูลต่าง ๆ เกี่ยวกับประเภทของอ็อบเจ็กต์ได้ เช่น:

ชื่อของคลาส (FullName): ชื่อคลาสพร้อม namespace

ชื่อของ assembly ที่คลาสนั้นอยู่ (Assembly)

คุณสมบัติและเมธอดของคลาส

สถานะของประเภท (เช่น ถ้าเป็นประเภทชนิด value type หรือ reference type)

9. ในภาษา C# การประกาศ class ไว้ก่อนหรือหลังการสร้าง object ต่างกันหรือไม่

*คำตอบ*

        ต่างกัน การประกาศ Class ต้องทำก่อนการสร้าง Object เสมอ เนื่องจาก Class เป็นโครงสร้างพื้นฐานที่ใช้กำหนดรูปแบบของ Object ดังนั้น ถ้าไม่มีการประกาศ Class ก่อน โปรแกรมจะไม่สามารถรู้จักประเภทของ Object ที่ต้องสร้าง และจะเกิดข้อผิดพลาดในระหว่างการคอมไพล์ (Compile-time Error)
